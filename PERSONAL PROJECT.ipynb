```python
from tkinter import *
import tkinter as tk
from tkinter import messagebox
import datetime as dt

root =  Tk()
root.title("TASK HELPER")
root.config()


Label(root, text='INPUT A TASK:-').grid(row=0)
Label(root, text='COMPLETED TASKS').grid(row=0, column=2)

# Creating label to display date
date = dt.datetime.now()
label_date= Label(root, text=f"{date:%A, %B %d, %Y}")
label_date.grid(row=1, column=0) 

ent = Entry(root)
ent.grid(row=0, column=1)
#Create a listbox and attach to window


def remove_item():
    selected_checkboxs = listbox.curselection()
    for selected_checkbox in selected_checkboxs[::-1]:
        listbox.delete(selected_checkbox)
        completed_listbox.delete(selected_checkbox)

def edit():
    task = ent.get()
    selected_checkboxs = listbox.curselection()
    edited_task = listbox.get(selected_checkboxs)
    listbox.delete(selected_checkboxs)
    listbox.insert(END, task)


def completed():
    selected_checkbox = listbox.curselection()
    completed_task = listbox.get(selected_checkbox)
    messagebox.showwarning("NOTE", "You can't delete tasks once they're added to completed tasks")
    confirm = messagebox.askyesno("COMPLETED TASKS", "Are you sure you have completed the following task?") 
    if confirm:
        listbox.delete(selected_checkbox)
        completed_listbox.insert(END, completed_task)

     
       
        
def add_item():
    task = ent.get()
    if task:
        listbox.insert(END, task)
        ent.delete(0, END)
        
        
        
        
                
        
listbox = Listbox(root, selectmode=BROWSE)
# Adding listbox to the left side of root window
listbox.grid(row=1, column=1)

completed_listbox = Listbox(root, selectmode=SINGLE, bg="green", relief="raised")
completed_listbox.grid(row=1, column=2)


 
    
Button(root,text="Delete", command=remove_item).grid(row=2, column=0)
Button(root,text="Add", command=add_item).grid(row=2, column= 1)
Button(root,text="Mark as completed", command=completed).grid(row=2, column=2)
Button(root,text="Edit", command=edit).grid(row=2, column=3)




root.mainloop()

```
